**[2.1] [5 marks]** Define a Python function called $mapping$(x) that, given a letter-list $x_L$, generates a number-list $x_N$ by mapping each of the 4 letters into a different number. Specifically, implement the changes

**[2.2] [5 marks]** Implement a function called $walker$(x) that, given a list $x$ of N numbers [x(0), x(1), x(2),...,x(N-1)], outputs a "walk list" $y=[y(0), y(1), ..., y(N)]$, defined recursively as: 
$$y(0) = 0,\\
y(n) = y(n-1) + x(n-1), \ \text{if} \ 0<n\leq N$$

---


````python
def walker(x):
    walk_list=[0]
#walk_list is initialised at 0 because we are told y(0) = 0
    for n in range(1,len(x)+1):
#I chose to iterate in the range(1,len(x)+1) because we were already given the first term of the walk list
        y = walk_list[n-1]+x[n-1]
#This adds the previous term of walk_list to the previous term of x to give the next term in the walk list
        walk_list.append(y)
    return walk_list
````

**[2.3] [10 marks]** Given points $(x_i,y_i)$ in the plane, a least-squares fit to a line $y=a x + b$ gives formulas for the slope $a$ and the intercept $y=b$ as
$$a=\frac{\langle x_iy_i\rangle-\langle x_i\rangle\langle y_i\rangle}{\langle x_i^2\rangle-\langle x_i\rangle^2}\quad\text{and}\quad b=\langle y_i\rangle-a\langle x_i\rangle\;,$$
where $\langle r_i\rangle$ denotes the average of the numbers $r_i$.

Without using any imported module, define a function called *linear_fit()* that:
* takes a (finite) list of points $z=[(x_0,y_0),(x_1,y_1),(x_2,y_2),...]$ as an input,
* fits a straight line to $y=a x + b$ by performing a least-squares fit,
* returns the values of $a$ and $b$.

Use
```python
def linear_fit_test(z):
    a,b=np.polyfit(*zip(*z), 1)
    return a,b
```

to compare the output of both functions for some well-chosen list of points to ensure that your function works appropriately.

---


````python
def linear_fit(z):
    x = [p[0] for p in z] #This defines the x coordinates of our data 
    y = [p[1] for p in z] #This defines the y coordinates of our data 
    x2 = [(p[0])**2 for p in z] #This takes all the x coordinates and squares them
    xy = [p[0]*p[1] for p in z] #This multiplies the x coordinates by the corresponding y coordinates
    
    average_of_xy = sum(xy)/len(z) 
    average_of_x = sum(x)/len(z)
    average_of_y = sum(y)/len(z)
    average_of_x2 = sum(x2)/len(z) 
    
#I then calculated all the averages I would need to calculate a and b
   
    a = ((average_of_xy)-((average_of_x)*(average_of_y)))/((average_of_x2)-(average_of_x)**2)
    b = (average_of_y)-a*(average_of_x)     
    
#I used the equations given in the question to then return a and b
    return a,b
````

For instance, in the sequence AACTGC we can find five different 2-grams: AA, AC, CT, TG and GC (notice that two consecutive 2-grams overlap in one letter). **[3.3] [10 marks]** Let $N(ij)$ be the frequency (that is, the number of occurrences) of the 2-gram 'ij' in the virus sequence, for $i,j=A,C,G,T$.
The transition matrix ${\bf T}=\{T_{ij}\}$ of a given sequence is defined such that the general term $T_{ij} = N(ij)/N_{tot}(i),$
where $N_{tot}(i)$ is the frequency of letter 'i' showing up in the sequence. By construction, all rows of $\bf T$ should sum up one.

Compute the $4\times 4$ transition matrix $\bf T$ of the virus sequence. Print this matrix and display it as a heatmap of this matrix using *seaborn*.

Confirm that the largest eigenvalue of $\bf T$ is one, and give the associated eigenvector $v$. Check your results by computing ${\bf T}v$.

---

````python
T = np.array(freq2(pairs)).reshape(4,4) 

#I changed T (a list of frequencies of each 2 gram) to a 4x4 matrix because we are told to compute a 4x4 transition matrix

T = T.astype(float) 

#T needs to be a floating point because the values we get are less than 1 and displaying them as integers would result in the zero matrix

for r in range(4):
    T[r] = T[r]/sum(T[r])
    
#T[r] represents the frequency of the 2-gram 'ij' in the virus sequence for i,j = A,C,G,T
#sum(T[r]) gives us the frequency of 'i' in the sequence


seaborn.heatmap(T,xticklabels = ['A','C','G','T'], yticklabels = ['A','C','G','T'])
plt.title('Heatmap of transition matrix', fontsize = 13)
plt.show()

eigenvalues, eigenvectors = la.eig(T)
v = eigenvectors[:,0]

#Each column vector of 'eigenvectors' represents an eigenvector
#eigenvectors[:,0] gives us the eigenvector corresponding to the eigenvalue of 1 since this is the first eigenvalue in the list returned to us by python
#Similarly, eigenvectors[:,1] would give the eigenvector corresponding to the second eigenvalue

print('\033[1m' + 'Eigenvalues:' + '\033[0m')
print()
print(eigenvalues)
print()
print('It is clear to see 1 is the largest eigenvalue')
print()
print('\033[1m' + 'Transition matrix of virus sequence:' + '\033[0m')
print()
print(T)
print()
print('\033[1m' + 'Assosciated eigenvector of the eigenvalue 1:' + '\033[0m')
print()
print(v)
print()
print('\033[1m' + 'Computation of Tv:' + '\033[0m')
print()
print(T@v)
print()
print('Since Tv = Î»v, where Î» is equal to 1 in this case, we can conclude the eigenvector (v) we have is correct')

# Tv is equal to v because the eigenvalue corresponding to this eigevector is 1. This has been explained with the statements I have printed
````

**[3.4] [8 marks]** Define a function called $deviation(x)$. The input of the function is a walk list $x$. This function performs the following computation:
* It iteratively considers all possible sublists $x_k=[x(0), ..., x(k)]$, where $k=1,2,4,\ldots$ runs through powers of $2$.
* For each possible value of $k$, the function computes $\Delta(k) = max(x_k) - min(x_k)$. 
* The function $deviation$ finally returns a list of the form $[(k,\Delta(k))]$ for all possible values of $k$. 
* That final list is called the *scaling* (this is important for next questions below)

For the case $x=[0,1,2,3,4,5,6,7,8]$ you should get $[(1, 1), (2, 2), (4, 4), (8, 8)]$, whereas for the case $x=[1,1,1,1]$ you should get $[(1,0),(2,0)]$.

---

````python
k_copy = []

def deviation(x):
    index = 0
    k = []
    delta_k = []

#The list 'k' gives all the possible values of k depending on the choice of x
#I defined a copy of it outside the function because I used it later in the project
#The largest value of k must be strictly less than the number of elements in the list and we know this from information given in the question

    while 2**index < len(x):
        k.append(2**index) #This creates our list of k values
        delta_k.append(max(x[:2**index+1])-min(x[:2**index+1])) #This creates our list of Î”(ð‘˜) values
        k_copy.append(2**index)
        index += 1

#index +=1 allows the while loop to keep iterating until all possible values of k are given (while calculating Î”(ð‘˜) too)
    
    new_list = list(zip(k,delta_k))
    return(new_list)

#list(zip(k,delta_k)) gives us a list in the form [(ð‘˜,Î”(ð‘˜))], as required in the question
 
#Testing     
print(deviation([0,1,2,3,4,5,6,7,8]))
print(deviation([1,1,1,1]))
````
