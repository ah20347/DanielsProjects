```python
def empty_entry(su):
    for i in range(len(su)):
        for j in range(len(su)):
            if su[i][j] == 0:
                return (i, j)
            
    return False
            
# I then decided to write a function that determines the positions of empty entries
```

```python
def valid(su, num, pos):
    # This checks the row 
    for i in range(len(su)):
        if su[pos[0]][i] == num and pos[1] != i:
            return False

    # This checks the column
    for i in range(len(su)):
        if su[i][pos[1]] == num and pos[0] != i:
            return False

    # This checks each box
    box_1 = pos[1] // 3
    box_2 = pos[0] // 3

    for i in range(box_2*3, box_2*3 + 3):
        for j in range(box_1 * 3, box_1*3 + 3):
            if su[i][j] == num and (i,j) != pos:
                return False

    return True

# This function checks to see whether we have a valid sudoku problem to be solved 
```
